### 1. Customer Demographics & Behavior

1. **Age-wise Sales Analysis**  
   - **Scenario:** You want to know which age group is buying the most during Diwali.  
   - **Question:** Which age group contributes the highest revenue, and how does purchase frequency vary by age group?

```python
# Count of orders per age group
age_group_counts = df["Age_Group"].value_counts()
#print(age_group_counts)

# Total revenue per age group
age_group_revenue = df.groupby("Age_Group")["Amount"].sum().sort_values(ascending=False)
#print(age_group_revenue)

# Visualization
plt.figure(figsize=(8,5))
# sns.barplot(x=age_group_revenue.index, y=age_group_revenue.values, palette="viridis")
sns.barplot(x=age_group_revenue.index, y=age_group_revenue.values, color="skyblue")
plt.title("Revenue by Age Group during Diwali")
plt.ylabel("Total Revenue")
plt.xlabel("Age Group")
plt.show()
```

2. **Gender-based Purchasing Patterns**  
   - **Scenario:** The marketing team wants to target customers by gender.  
   - **Question:** What is the total revenue from male vs female customers? Which gender prefers which product category?

```python
# Total revenue per age group
gender_wise_revenue = df.groupby("Gender")["Amount"].sum()
# gender_wise_revenue

# Visualization
plt.figure(figsize=(8,5))
plt.pie(
    gender_wise_revenue.values,
    labels=gender_wise_revenue.index,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Revenue by Gender during Diwali")
plt.show()
```

3. **Marital Status Impact**  
   - **Scenario:** Sales team wants to know if marital status affects buying patterns.  
   - **Question:** Do married customers spend more than unmarried ones? Which categories are preferred by married vs unmarried customers?

```python
# Total spending by marital status
marital_spend = df.groupby("Marital_Status")["Amount"].sum()
# marital_spend

# Visualization
plt.figure(figsize=(6,4))
sns.barplot(
    x=marital_spend.index,
    y=marital_spend.values
)
plt.title("Total Spending by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Total Amount")
plt.show()
```

4. **Occupation vs Purchase Behavior**  
   - **Scenario:** The business wants to run targeted campaigns for different occupations.  
   - **Question:** Which occupations contribute most to revenue? Which product categories are popular among students, professionals, or homemakers?

```python
# Total spending by Occupation
Occupation_Spend = (
    df.groupby("Occupation")["Amount"]
    .sum()
    .sort_values(ascending=False)
)

# Visualization
plt.figure(figsize=(15,5))
sns.barplot(
    x=Occupation_Spend.index,
    y=Occupation_Spend.values
)
plt.title("Total Revenue by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Total Amount")
plt.xticks(rotation=45)
plt.show()
```

5. **Customer Loyalty Analysis**  
   - **Scenario:** Identify repeat customers to offer them loyalty discounts.  
   - **Question:** Which customers purchased more than once during Diwali? How much revenue do repeat customers generate compared to one-time buyers?



```python
# Aggregate at customer level
customer_orders = (
    df.groupby("User_ID")
    .agg(
        total_orders=("Orders", "sum"),
        total_spend=("Amount", "sum")
    )
    .reset_index()
)

# Classify customers
customer_orders["Customer_Type"] = customer_orders["total_orders"].apply(
    lambda x: "Repeat Customer" if x > 1 else "One-Time Buyer"
)

# Revenue comparison
revenue_by_type = (
    customer_orders.groupby("Customer_Type")["total_spend"]
    .sum()
    .reset_index()
    .sort_values("total_spend", ascending=False)
)

# Visualization
plt.figure(figsize=(6,4))
ax = sns.barplot(
    data=revenue_by_type,
    x="Customer_Type",
    y="total_spend"
)

for p in ax.patches:
    ax.annotate(
        f'{p.get_height():,.0f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom'
    )

plt.title("Revenue: Repeat vs One-Time Customers")
plt.xlabel("Customer Type")
plt.ylabel("Total Revenue")
plt.show()

# Customer count
customer_orders["Customer_Type"].value_counts()
```

8. **Price vs Quantity Analysis**  
   - **Scenario:** To plan discounts, you need to see the relationship between product price and quantity sold.  
   - **Question:** Are high-priced items selling less than low-priced items? What is the correlation between price and quantity sold?

```python
# Aggregate at product level
product_sales = (
    df.groupby("Product_ID")
    .agg(
        total_quantity=("Orders", "sum"),
        total_revenue=("Amount", "sum")
    )
    .reset_index()
)

# Derive unit price
product_sales["Unit_Price"] = (
    product_sales["total_revenue"] / product_sales["total_quantity"]
)

product_sales.head()

# Visualization
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=product_sales,
    x="Unit_Price",
    y="total_quantity"
)

plt.title("Price vs Quantity Sold")
plt.xlabel("Unit Price")
plt.ylabel("Total Quantity Sold")
plt.show()


correlation = product_sales["Unit_Price"].corr(
    product_sales["total_quantity"]
)

correlation
```

9. **Discount Impact**  
    - **Scenario:** Marketing team wants to measure effectiveness of discounts.  
    - **Question:** Do products with discounts sell significantly more than products without discounts? How does discount percentage affect revenue?

```python
# Aggregate at product level
product_sales = (
    df.groupby("Product_ID")
    .agg(
        total_quantity=("Orders", "sum"),
        total_revenue=("Amount", "sum")
    )
    .reset_index()
)

# Derive unit price
product_sales["Unit_Price"] = (
    product_sales["total_revenue"] / product_sales["total_quantity"]
)

# Split products into Low-price vs High-price (proxy for discount)
median_price = product_sales["Unit_Price"].median()

product_sales["Price_Type"] = product_sales["Unit_Price"].apply(
    lambda x: "Lower Priced (Discounted Proxy)" if x <= median_price else "Higher Priced"
)

product_sales.head()

quantity_comparison = (
    product_sales.groupby("Price_Type")["total_quantity"]
    .mean()
    .reset_index()
)

quantity_comparison

# Visualization
plt.figure(figsize=(6,4))
sns.barplot(
    data=quantity_comparison,
    x="Price_Type",
    y="total_quantity"
)
plt.title("Quantity Sold: Lower vs Higher Priced Products (Proxy)")
plt.xlabel("Product Type")
plt.ylabel("Average Quantity Sold")
plt.show()
```

### 4. Advanced Analytical Scenarios

15. **Customer Segmentation for Marketing**  
    - **Scenario:** Targeted marketing campaigns for high-value customers.  
    - **Question:** Segment customers into high, medium, and low spenders. Which segment contributes most to revenue?

```python
# Calculate total spending per customer
customer_spend = (
    df.groupby("User_ID")["Amount"]
    .sum()
    .reset_index()
)

# Segment customers into Low, Medium, and High spenders
customer_spend["Spend_Segment"] = pd.qcut(
    customer_spend["Amount"],
    q=3,
    labels=["Low Spender", "Medium Spender", "High Spender"]
)

customer_spend.head()

# Calculate total revenue contributed by each segment
segment_revenue = (
    customer_spend
    .groupby("Spend_Segment", observed=True)["Amount"]
    .sum()
    .reset_index()
)

segment_revenue

# Visualization: Revenue contribution by customer segment
plt.figure(figsize=(6,4))
sns.barplot(
    data=segment_revenue,
    x="Spend_Segment",
    y="Amount"
)
plt.title("Revenue Contribution by Customer Segment")
plt.xlabel("Customer Segment")
plt.ylabel("Total Revenue")
plt.show()
```

    - **Scenario:** Identify upselling opportunities.  
    - **Question:** Which customers buy multiple categories? Are there patterns in cross-category purchases?

```python
# Count unique product categories purchased by each customer
customer_categories = (
    df.groupby("User_ID")["Product_Category"]
    .nunique()
    .reset_index(name="category_count")
)

customer_categories.head()


# Identify customers who purchased from more than one category
multi_category_customers = customer_categories[
    customer_categories["category_count"] > 1
]

multi_category_customers.head()


# Merge category count back into the main dataframe
df_with_category_count = df.merge(
    customer_categories,
    on="User_ID",
    how="left"
)

# Label customers based on number of categories purchased
df_with_category_count["Customer_Type"] = df_with_category_count["category_count"].apply(
    lambda x: "Multi-Category Buyer" if x > 1 else "Single-Category Buyer"
)

# Calculate total revenue by customer type
cross_category_revenue = (
    df_with_category_count
    .groupby("Customer_Type", observed=True)["Amount"]
    .sum()
    .reset_index()
)

cross_category_revenue


# Visualization: Revenue comparison between customer types
plt.figure(figsize=(6,4))
sns.barplot(
    data=cross_category_revenue,
    x="Customer_Type",
    y="Amount"
)
plt.title("Revenue: Multi-Category vs Single-Category Buyers")
plt.xlabel("Customer Type")
plt.ylabel("Total Revenue")
plt.show()


# Create unique customerâ€“category combinations
customer_category_pairs = (
    df[["User_ID", "Product_Category"]]
    .drop_duplicates()
)

# Generate category pairs purchased by the same customer
category_pairs = (
    customer_category_pairs
    .merge(
        customer_category_pairs,
        on="User_ID"
    )
)

# Remove same-category combinations
category_pairs = category_pairs[
    category_pairs["Product_Category_x"] < category_pairs["Product_Category_y"]
]

# Count frequency of each category pair
top_category_pairs = (
    category_pairs
    .groupby(["Product_Category_x", "Product_Category_y"])
    .size()
    .reset_index(name="customer_count")
    .sort_values("customer_count", ascending=False)
)

top_category_pairs.head(10)
```

18. **Revenue Forecasting**  
    - **Scenario:** Predicting future Diwali sales based on historical trends.  
    - **Question:** Can we predict total revenue for the next Diwali season using past sales trends?

```python
# Calculate total revenue from historical Diwali sales
total_revenue = df["Amount"].sum()
total_revenue


# Assume expected growth rate for next Diwali season
assumed_growth_rate = 0.12   # 12% expected growth

# Predict next Diwali revenue using growth assumption
predicted_revenue_next_diwali = total_revenue * (1 + assumed_growth_rate)
predicted_revenue_next_diwali


# Create a summary table for forecasting results
forecast_df = pd.DataFrame({
    "Metric": [
        "Historical Diwali Revenue",
        "Expected Growth Rate",
        "Predicted Next Diwali Revenue"
    ],
    "Value": [
        total_revenue,
        "12%",
        predicted_revenue_next_diwali
    ]
})

forecast_df
```

19. **Product Performance Heatmap**  
    - **Scenario:** Visualizing performance for quick insights.  
    - **Question:** Create a heatmap showing revenue of product categories across different states.

```python
# Create a pivot table for heatmap
# Rows represent States
# Columns represent Product Categories
# Values represent total revenue

heatmap_data = (
    df.pivot_table(
        index="State",
        columns="Product_Category",
        values="Amount",
        aggfunc="sum"
    )
)

heatmap_data.head()


# Visualization: Heatmap of product category revenue across states
plt.figure(figsize=(14,8))

sns.heatmap(
    heatmap_data,
    cmap="YlOrRd",
    linewidths=0.5
)

plt.title("Product Category Revenue Across States")
plt.xlabel("Product Category")
plt.ylabel("State")
plt.show()
```

20. **High-Value Customer Targeting**  
    - **Scenario:** Increase loyalty among top customers.  
    - **Question:** Who are the top 5% of customers contributing to revenue, and what are their buying habits?

```python
# STEP 1: Calculate total revenue per customer
customer_spend = (
    df.groupby("User_ID")["Amount"]
    .sum()
    .reset_index()
)

# STEP 2: Identify top 5% high-value customers
top_5_percent_threshold = customer_spend["Amount"].quantile(0.95)

high_value_customers = customer_spend[
    customer_spend["Amount"] >= top_5_percent_threshold
]

# STEP 3: Revenue contribution of top 5%
top_5_revenue = high_value_customers["Amount"].sum()
total_revenue = customer_spend["Amount"].sum()

top_5_revenue, total_revenue

# STEP 4: Merge high-value customers with main dataset
high_value_df = df.merge(
    high_value_customers[["User_ID"]],
    on="User_ID",
    how="inner"
)

# STEP 5: Product category preference
top_categories = (
    high_value_df.groupby("Product_Category")["Amount"]
    .sum()
    .reset_index()
    .sort_values("Amount", ascending=False)
)

# STEP 6: Age group contribution
high_value_age = (
    high_value_df.groupby("Age_Group")["Amount"]
    .sum()
    .reset_index()
    .sort_values("Amount", ascending=False)
)

# STEP 7: Occupation contribution
high_value_occupation = (
    high_value_df.groupby("Occupation")["Amount"]
    .sum()
    .reset_index()
    .sort_values("Amount", ascending=False)
)

# STEP 8: Visualization â€“ Top product categories
plt.figure(figsize=(8,4))
sns.barplot(
    data=top_categories.head(10),
    x="Product_Category",
    y="Amount"
)
plt.title("Top Product Categories Purchased by High-Value Customers")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```
