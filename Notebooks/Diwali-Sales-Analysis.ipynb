## 4. Data Visualization

**What it is:**  
The process of creating **graphs, charts, and plots** to visually represent data using libraries like **matplotlib** and **seaborn**.

### 1. Customer Demographics & Behavior

1. **Age-wise Sales Analysis**  
   - **Scenario:** You want to know which age group is buying the most during Diwali.  
   - **Question:** Which age group contributes the highest revenue, and how does purchase frequency vary by age group?

```python
# Count of orders per age group
age_group_counts = df["Age_Group"].value_counts()
#print(age_group_counts)

# Total revenue per age group
age_group_revenue = df.groupby("Age_Group")["Amount"].sum().sort_values(ascending=False)
#print(age_group_revenue)

# Visualization
plt.figure(figsize=(8,5))
# sns.barplot(x=age_group_revenue.index, y=age_group_revenue.values, palette="viridis")
sns.barplot(x=age_group_revenue.index, y=age_group_revenue.values, color="skyblue")
plt.title("Revenue by Age Group during Diwali")
plt.ylabel("Total Revenue")
plt.xlabel("Age Group")
plt.show()
```
